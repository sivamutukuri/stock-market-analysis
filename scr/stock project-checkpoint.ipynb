•What was the moving average of various stocks?


```python
data1=yf.download("AAPL",start="2022-02-03",end="2023-02-03")
```

```python
data1
```

```python
mavg=[20,50,100]
for MA in mavg:
    column_name = f"MA for {MA} days" 
    data1[column_name]=data1['Adj Close'].rolling(MA).mean()
```

```python
data1
```

• Add new column ‘Trend’ whose values are based on the 'Daily Return'.


```python
data1['Trend'] = 'Stable'  # Default value for 'Trend'
data1.loc[data2['dailyReturns'] > 0, 'Trend'] = 'Up'       # Positive 'Daily Return' indicates an upward trend
data1.loc[data2['dailyReturns'] < 0, 'Trend'] = 'Down'
```
